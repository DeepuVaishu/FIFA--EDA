# Exploratory Data Analysis on FIFA Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Cleaning And Preprocessing

```python
data = pd.read_csv("Downloads/fifa_data.csv")
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
missing_values = data.isnull().sum()
print("\nMissing Values in Each Column:")
print(missing_values[missing_values > 0])
```

```python
num_columns = data.select_dtypes(include=['float64', 'int64']).columns
for col in num_columns:
    if data[col].isnull().sum() > 0:
        median_value = data[col].median()
        data.fillna({col:median_value}, inplace=True)   
```

```python
duplicates = data.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")
```

```python
data.columns = data.columns.str.lower().str.replace(' ', '_')
print("\nStandardized Column Names:")
print(data.columns)
```

```python
cat_columns = data.select_dtypes(include=['object']).columns
for col in cat_columns:
    if data[col].isnull().sum() > 0:
        mode_value = data[col].mode()[0]
        data.fillna({col:mode_value}, inplace=True)
```

```python
print(data.isnull().sum().sum())
```

### 4. Distribution of Player Wages

```python
data['wage'].head()
```

```python
# Convert 'wage' column to numeric (handling errors)
# Remove the € sign and convert the wage column to integers
data['wage'] = data['wage'].str.replace('€', '').str.replace('K', '000').astype(float)

# Define new wage bins and labels based on the updated wage ranges
wage_bins = [0, 5000, 50000, 200000, data['wage'].max() + 1]
wage_labels = ['Low (< 5K)', 'Medium (5K - 50K)', 'High (50K - 200K)', 'Very High (> 200K)']

# Create 'wage_group' column
data['wage_group'] = pd.cut(data['wage'], bins=wage_bins, labels=wage_labels, right=False)

# Display count of players in each wage group
print(data['wage_group'].value_counts())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['wage_group'], kde=True, color='red')
plt.title('Distribution of Player Wages', fontsize=16)
plt.xlabel('Wage (in Euros)')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

### 3. Player Attributes by Position

```python
data['position'].head()
```

```python
# Define common attributes of interest
attributes = ['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']
existing_attributes = [attr for attr in attributes if attr in data.columns]

# Check if 'position' column and attributes are present
if 'position' in data.columns and existing_attributes:
    # Grouping positions into broader categories
    position_map = {
        'GK': 'Goalkeeper',
        'CB': 'Defender', 'LB': 'Defender', 'RB': 'Defender',
        'LCB': 'Defender', 'RCB': 'Defender', 'LWB': 'Defender', 'RWB': 'Defender',
        'CDM': 'Midfielder', 'CM': 'Midfielder', 'LCM': 'Midfielder', 'RCM': 'Midfielder',
        'CAM': 'Midfielder', 'LM': 'Midfielder', 'RM': 'Midfielder',
        'LW': 'Forward', 'RW': 'Forward', 'ST': 'Forward', 'CF': 'Forward',
        'LS': 'Forward', 'RS': 'Forward', 'RF': 'Forward', 'LF': 'Forward'
    }

    # Create a new column for grouped positions
    data['position_group'] = data['position'].map(position_map).fillna('Other')

    # Melting the dataframe with a unique value name
    melted_df = data.melt(
        id_vars=['position_group'],
        value_vars=existing_attributes,
        var_name='attribute',
        value_name='attribute_score'  # Renamed to avoid conflict
    )
    
    plt.figure(figsize=(14, 8))
    sns.boxplot(data=melted_df, x='attribute', y='attribute_score', hue='position_group', palette='Set2')
    plt.title('Player Attributes by Position Group', fontsize=16,color='blue')
    plt.xlabel('Attribute',fontsize=12,color='navy')
    plt.ylabel('Score',fontsize=12,color='navy')
    plt.legend(title='Position Group')
    plt.show()
else:
    print("Required columns for attributes by position analysis are missing.")
```

## Correlation Analysis

```python
# Function to clean and convert monetary values
def convert_monetary_value(value):
  # Handle cases where value is already a float (no conversion needed)
  if isinstance(value, float):
    return value

  # Remove the € sign (if present)
  value = value.strip('€')  # This removes leading/trailing € signs

  # Convert K to thousands and M to millions
  if 'K' in value:
    return float(value.replace('K', '')) * 1_000
  elif 'M' in value:
    return float(value.replace('M', '')) * 1_000_000
  else:
    # Assume value is a plain number without suffix
    return float(value)
      
# Apply the conversion function to 'wage' and 'value' columns
data['wages'] = data['wage'].apply(convert_monetary_value)
data['value_clean'] = data['value'].apply(convert_monetary_value)

# Check the cleaned columns
print(data[['wages', 'value_clean']].head())
```

```python
# Calculate the correlation matrix
corr_matrix = data[['age', 'overall', 'potential', 'wages', 'value_clean']].corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of FIFA Dataset Features', fontsize=16,color='red')
plt.show()
```

## Conclusion


In this exploratory data analysis of the FIFA dataset, we examined key attributes of football players, including their overall ratings, potential, wages, and market values. Our goal was to uncover insights into player performance, financial trends, and club dynamics.

**Key Insights**:
- **Position Analysis**: Central Midfielders (CM) and Strikers (ST) often have higher average overall ratings, indicating their pivotal roles in gameplay.
- **Wage Group Distribution**: Most players fell into the 'Medium' wage group (5K - 50K), with a smaller segment in the 'Very High' category (> 200K), highlighting the disparity in player wages.
- **Top Players**: The highest-rated players were typically associated with top clubs like FC Barcelona and Real Madrid, known for their investment in elite talent.
- **Club Analysis**: Clubs like Paris Saint-Germain and Manchester City stood out for their high average player ratings and substantial wage bills, reflecting their financial strength and focus on acquiring top-tier players.

**Challenges and Limitations**:
A notable challenge was handling inconsistent formats in financial data (e.g., wages and market values expressed with currency symbols). Additionally, the dataset might have biases as it predominantly features popular clubs and players from major leagues.

**Future Work**:
Future analysis could explore time-series trends in player ratings or wages over multiple seasons. Additionally, integrating data from external sources like match statistics could provide a more comprehensive view of player performance.

**Final Thoughts**:
This EDA provided valuable insights into player attributes and club dynamics in FIFA. Understanding these factors can help clubs make better scouting decisions, and enthusiasts can gain deeper knowledge of football player statistics.
